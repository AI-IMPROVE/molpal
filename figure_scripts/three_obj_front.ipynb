```python
import pickle 
from pathlib import Path 

base_dir = Path('enamine_IGF1R_EGFR_01')

with open(base_dir/'exhaustive_metrics_0.001.pickle', 'rb') as f: 
    res = pickle.load(f)

points = res['true_front_points']
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from plot_utils import set_size, set_style
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator

set_style()
mpl.rcParams['xtick.major.pad']=0
mpl.rcParams['grid.linewidth'] = 0.2
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_trisurf( 
    points[:,0], points[:,1], points[:,2],  
    linewidth=0.1, edgecolor=mpl.colors.colorConverter.to_rgba('black', alpha=.7),
    alpha=0.8, cmap=plt.cm.cividis)
ax.scatter(points[:,0], points[:,1], points[:,2], color='black', linewidth=0, s=5)

ax.set(xlabel=('-IGF1R'), ylabel=('-EGFR'), zlabel=('CYP3A4'))
# fig.colorbar(surf, shrink=0.4)

# ax.zaxis.draw(fig)

# ax.invert_yaxis()
# ax.invert_xaxis()

ax.view_init(20,70)
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.zaxis.set_major_locator(MaxNLocator(integer=True))
ax.xaxis.set_major_locator(MaxNLocator(integer=True))


set_size(2,2,ax=ax)
fig.tight_layout()
plt.show()


fig.savefig(base_dir/f'three_obj_pf.pdf',bbox_inches='tight', dpi=200, transparent=True, pad_inches=0.5)



```
